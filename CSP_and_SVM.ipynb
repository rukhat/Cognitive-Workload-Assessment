```python
import mne
from glob import glob
import numpy as np
from mne.decoding import CSP
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
zeroBACK_files = glob('../epoch_data/N-Back/sub-02/*zeroBACK.set')
oneBACK_files = glob('../epoch_data/N-Back/sub-02/*oneBACK.set')
twoBACK_files = glob('../epoch_data/N-Back/sub-02/*twoBACK.set')
```

```python
zeroBACK_epochs = []
oneBACK_epochs = []
twoBACK_epochs = []

for zero in zeroBACK_files:
    zeroBACK_epoch = mne.io.read_epochs_eeglab(zero)
    print(zeroBACK_epoch.get_data().shape)
    zeroBACK_epochs.append(zeroBACK_epoch)

for one in oneBACK_files:
    oneBACK_epoch = mne.io.read_epochs_eeglab(one)
    print(oneBACK_epoch.get_data().shape)
    oneBACK_epochs.append(oneBACK_epoch)

for two in twoBACK_files:
    twoBACK_epoch = mne.io.read_epochs_eeglab(two)
    print(twoBACK_epoch.get_data().shape)
    twoBACK_epochs.append(twoBACK_epoch)
```

```python
csp = CSP(n_components=4, reg=None, log=True, norm_trace=False)
ss = StandardScaler()

zeroBack_features = []
oneBack_features = []
twoBack_features = []

for zero in zeroBACK_epochs:
    zeroBack_features.append(ss.fit_transform(csp.fit_transform(zero.get_data(), zero.events[:, -1])))
    # zeroBack_labels.append(zero.events[:, -1])

for one in oneBACK_epochs:
    oneBack_features.append(ss.fit_transform(csp.fit_transform(one.get_data(), one.events[:, -1])))
    # oneBack_labels.append(one.events[:, -1])

for two in twoBACK_epochs:
    twoBack_features.append(ss.fit_transform(csp.fit_transform(two.get_data(), two.events[:, -1])))
    # twoBack_labels.append(two.events[:, -1])

zeroBack_labels = [0] * len(zeroBack_features[0]) * 3
oneBack_labels = [1] * len(oneBack_features[0]) * 3
twoBack_labels = [2] * len(twoBack_features[0]) * 3
# np.save('zerobackFeatures.npy', zeroBack_features, allow_pickle=True)


# oneBack_features = ss.fit_transform(csp.fit_transform(oneBACK_epoch.get_data(), oneBACK_epoch.events[:, -1]))
# oneBack_labels = [1]*len(oneBack_features)
# oneBack_labels = oneBACK_epoch.events[:, -1]
# np.save('onebackFeatures.npy', oneBack_features, allow_pickle=True)

# twoBack_features = ss.fit_transform(csp.fit_transform(twoBACK_epoch.get_data(), twoBACK_epoch.events[:, -1]))
# twoBack_labels = [2]*len(twoBack_features)
# twoBack_labels = twoBACK_epoch.events[:, -1]
# np.save('onebackFeatures.npy', oneBack_features, allow_pickle=True)
```

```python
csp = CSP(n_components=4, reg=None, log=True, norm_trace=False)
ss = StandardScaler()
```

```python
zeroBack = np.concatenate((zeroBack_features[0], zeroBack_features[1], zeroBack_features[2]), axis=0)
oneBack = np.concatenate((oneBack_features[0], zeroBack_features[1], oneBack_features[2]), axis=0)
twoBack = np.concatenate((twoBack_features[0], twoBack_features[1], twoBack_features[2]), axis=0)
```

```python
total_features = np.concatenate((zeroBack, oneBack, twoBack), axis=0)
total_labels = zeroBack_labels + oneBack_labels + twoBack_labels

np.save('../feature_data/N-Back/sub-02/total_features.npy', total_features, allow_pickle=True)
np.save('../feature_data/N-Back/sub-02/total_labels.npy', total_labels, allow_pickle=True)
```

```python
total_features.shape
```

```python
clf = svm.SVC()

X_train, X_test, y_train, y_test = train_test_split(total_features, total_labels, test_size=0.2, shuffle=True)

clf.fit(X_train, y_train)
```

```python
pred = clf.predict(X_test)
print(accuracy_score(y_test, pred))
```
